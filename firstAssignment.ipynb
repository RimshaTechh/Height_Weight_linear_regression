```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
df = pd.read_csv("/content/drive/MyDrive/Data Sets/height_weight.csv")
df
plt.figure(figsize=(6,4))
plt.xlabel("Weight",color="green")
plt.ylabel("Height",color="green")
plt.scatter(df.Weight,df.Height)
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[['Weight']],df.Height)
```

```python
lr.predict([[56.2]])
print("R2 Score:", lr.score(df[['Weight']],df.Height))
print("Coefficient:", lr.coef_[0])
print("Intercept:", lr.intercept_)
```

```python
#Regression Line Plot
plt.xlabel("Weight",color="green")
plt.ylabel("Height",color="green")
plt.scatter(df.Weight,df.Height)
plt.plot(df.Weight,lr.predict(df[['Weight']]),color='red')
```
